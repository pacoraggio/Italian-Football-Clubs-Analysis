# UEFA club European Cup Web Scraping

In this notebook, we will develop and review a web scraper to extract all matches played in the major European Club Cup tournments from 1980 onward. The goal is to create a dataset with the following information:

- `season`
- `competition`
- `round`
- `team_a`,
- `country_a`,
- `team_b`,
- `country_b`
- `score_leg1`
- `score_leg2`
- `score_aggregate`
- `extra_time`
- `penalties`

The first step of our analysis will concentrate on single results only, meaning that we will use the information stored in the first nine columns only. The idea, at least as a first insight, is to compare the winning/points ratio of all clubs by nation without giving any weight to the competition, something that can be done in a further step. 

The process to build the webscraper is iterative and is done in collaboration with Claude. The web resources that will be used as main source is well structured and yet contains different spots where the script to scrape the data must be adapted as we collect the data - *learning point*.  

## Scraping data from kassiesa.net

- source: [kassiesa.net](https://kassiesa.net/uefa/index.html)

This database contains all match results and calculated coefficients of European Cup football since 1955, covering all rounds including qualifiers. It's one of the most thorough free sources and is well-regarded among football statisticians. It will be used as starting point for collecting the data. A possible exhaustive approach might be:

- Starting with Kassiesa.net as ground truth (it has all three competitions, nationalities, and correct results). Scrapeing the season-by-season match pages (they're structured HTML tables). This will give us ~1980–present across all competitions.
- Supplement with jalapic/champs CSV (free direct download) for Champions League 1980–2017 as a cross-check.
- For the extra-time / penalties distinction, ***Kassiesa.net does not always separate these from 90-min scores***. We may need to cross-reference with EuroCupsHistory (eurocupshistory.com) for knockout rounds specifically.
